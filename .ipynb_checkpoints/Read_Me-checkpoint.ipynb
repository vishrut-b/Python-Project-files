### Summary : In this project, I have built a Multi Order Coverage(MOC) library, and verified its effectiveness by comparing my result with the corresponding result using the mocpy library.

#### Catalog used = "II/7A/catalog"


#### I have taken the follwing steps: 

1. I extract a table in VOT format, and then I built an astropy table with an added HEALPix column in order = 4 (which can be modified). I used astropy-healpix library, to extract the HEALPix cells from RA and DEC. 

2. Then I create a simple HEALPix pixelation from the HEALPix column at order = 3(which can be modified) and put it in a text file "Serialised_text.txt"

3. Then I create a module called "mod.py", where I have defined several functions depending upon their intended purpose. 'details' is a simple function which basically returns the value of the order we choose and the ideal number of HEALPIX cells it should contain. 'current_order' also returns the value of the order we choose, along with the corresponding HEALPix cells extract from the HEALPix column of that order. The final result of this function is a list of sorted and unique HEALPix cells. The next function I define is called "increase_order", and it gives us the HEALPix cells of the order one step above the one we choose for the function. This function is not used in the final and the most import function "My_MOC", which builds the MOC tree, but I have kept it just for conceptual clarity about how the MOC actually works. Function "My_MOC" builds the MOC tree. 

4. I use the different functions listed in the module 'mod' to illustrate what they do, and I go on to plot my MOC using the serialisation produced by my function "My_MOC", and then also compare the plot with the plot generated by MOC.from_vizier_table function, in order to verify the accuracy of my code.

### Additional info: The Zip file contains my  notebook, this ReadMe file, the text file containing the text serialisation, and two plots, one of which is the plot generated by my function "My_MOC" and the other one being the plot generated by the function "MOC.from_vizier_table"
